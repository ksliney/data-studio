```python
#first let's look at the data for 2008, and then replicate for other years?
import matplotlib.font_manager
from IPython.core.display import HTML
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')
#note to self: a row is an axis and a column is an axis
```

```python
df1 = pd.read_csv("Beijing_2008_HourlyPM2.5_created20140325.csv", encoding='latin1')
```

```python
df1['QC Name'].astype(str)
df1.head()
```

```python
df1.tail()
```

```python
#remove the missing data and negative values for year 2008
new_df1 = df1[df1['QC Name'].str.contains("Missing") == False]
```

```python
####APRIL
```

```python
#let's look at the first month we have data for (april), and evenually get the average by month
april_df = new_df1.loc[new_df1['Month'] == 4] 
```

```python
april_avg = april_df['Value'].mean(axis=0)
april_avg
```

```python
april_df['Average_PM_value']=april_avg
```

```python
april_df.plot(x='Month', y='Average_PM_value', kind='scatter', title='Average Respirable Particulate Matter in April 2008', legend=True)
```

# Findings:

What I've done:
My question is how does Beijing's air quality currently compare to 2008, when the US Embassy began collecting the data. Using the US Embassy's published collection of hourly air quality data in Beijng, I made a scatterplot of the average respirable particulate matter by month. First I removed any missing values, and then I plotted the average PM reading for that month. Note that for 2008, the first year of collection,  there is no data for Jan-March and December.
    
What I've found:
Beijing appears to have less respirable particulate matter in the Fall.

```python
#now to work on styling 2008
#inspiration: https://fivethirtyeight.com/features/dont-be-fooled-by-baseballs-small-budget-success-stories/ 
ax = spring_df.plot(x='Month', y='Average_PM_value', kind='scatter', s=50, color='black', alpha=.2, title='Average Respirable Particulate Matter in 2008', legend=False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.tick_params(which='major', left=False, bottom=False, grid_alpha=0.5)
ax.set_facecolor('#e1e1e1')

#so the only grid lines that show up are x
ax.grid(True, color='white', linewidth=1)
ax.set_axisbelow(True)
ax.set_xlim(3,13)
ax.set_ylabel("PM 2.5 Value")
ax.yaxis.label.set_visible(True)
```
