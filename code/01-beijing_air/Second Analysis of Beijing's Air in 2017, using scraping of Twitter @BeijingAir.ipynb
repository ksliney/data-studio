```python
df = pd.read_csv("twitterair2017.csv", encoding='latin1')
df = df[df['tweet'].str.contains("No Data")==False]
df = pd.DataFrame(df[df['tweet'].str.contains("No Data")==False], columns = ['tweet'])
```

```python
df['date'] = df['tweet'].str.extract(r'(\d{2}-\d{2}-\d{4})')
df['2.5_reading'] = df['tweet'].str.extract(r'PM2.5*;\s(\d*.\d*)')
df = df.dropna()
df
```

```python
df['date'] = pd.to_datetime(df['date'])
df
```

```python
df2017 = df.groupby('date')['2.5_reading'].mean().reset_index()
df2017.to_csv('beijiing_air_2017_twitter.csv')
```
