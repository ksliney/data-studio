```python
import pandas as pd
import matplotlib.pyplot as plt
from cycler import cycler
import statistics 
%matplotlib inline
```

```python
df=pd.DataFrame()
df = pd.read_csv("obama_gender_edu_clean.csv")
```

```python
plt.rcParams.update({
'axes.prop_cycle': cycler('color', ['#C0C0C0', '#00a4dc', '#C0C0C0', '#C0C0C0', '#C0C0C0', '#C0C0C0', '#C0C0C0']),
'axes.labelsize': 'x-large',
'ytick.color': 'black',
'xtick.color': 'black',
'axes.grid': False,
'figure.facecolor': 'white'
})

df_grad = df['Masters'].dropna()
ax = df_grad.value_counts().plot(kind='barh', figsize=(10, 15))
ax.set_ybound(lower=-.5, upper=15.5)
plt.tick_params(labelsize = 15)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_title("Georgetown beats SAIS grad representation; Hopkins wins overall",loc='center',fontsize = 25, pad=20)
```

```python
plt.rcParams.update({
'figure.facecolor': 'white',
'grid.color': '#c8c8c9',
'axes.prop_cycle': cycler('color', ['#1690d3', '#f55486',]),
'axes.grid': False,
'font.serif' : 'EcoNew Head',
#size of 'State' label
'axes.labelsize': 'x-large',
'axes.titlesize':'x-large'
})


ax = df['gender'].value_counts().plot(kind='barh',figsize=(10, 5))
ax.grid(False, axis='y')
ax.set_title("Women v. Men on Obama's NSC", fontsize = 20, pad=30)
ax.xaxis.set_ticks_position('top')
plt.tick_params(labelsize = 13)
ax.tick_params(which='major', labelcolor='#1d1a1b')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
```
